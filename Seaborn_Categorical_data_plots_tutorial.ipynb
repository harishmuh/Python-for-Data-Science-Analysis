# **Categorical Data Plots**
___

Now let's discuss of how using seaborn library to plot categorical data! There are a few main plot types for this:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Let's go through examples of each!

**Importing seaborn**

```python
# Importing seaborn library
import seaborn as sns

# Importing matplotlib library (supporting display)
import matplotlib.pyplot as plt
%matplotlib inline

# Importing numpy
import numpy as np

# Customizing with 'ggplot' style
plt.style.use('ggplot')

# Importing warnings
import warnings
warnings.filterwarnings('ignore')
```

**Loading dataset**

```python
# Loading dataset 'tips' from seaborn library
df = sns.load_dataset('tips')
```

```python
# Quick review of top and bottom dataset
display(df.head(5))
```

## **Barplot**

 Barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
# Creating Barplot
sns.barplot(x='sex', y='total_bill', data=df, palette='Set1')

plt.title('Barplot of average total bill by sex')
plt.show()
```

```python
# Creating Barplot with customize unique color
sns.barplot(
    x='sex', y='total_bill', data=df,
    palette={'Male': '#1f77b4', 'Female': '#ff7f0e'}
)

plt.title('Barplot of average total bill by sex')
plt.show()
```

We can change the estimator object to your own function, that converts a vector to a scalar:

```python
# Creating Barplot with modified estimator object (e.g., using standard deviation)
sns.barplot(x='sex',y='total_bill',data=df,
            palette='Set1', estimator=np.std)

plt.title('Barplot of average total bill by sex')
plt.show()
```

## **Countplot**

This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value:

```python
# Creating countplot
sns.countplot(x='sex', data=df, palette='Set1')

plt.title('Countplot of sex')
plt.show()
```

## **Boxplot**

boxplots  are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers using a method that is a function of the inter-quartile range.

```python
# Creating Boxplot
sns.boxplot(x="day", y="total_bill", data=df, palette='rainbow')

plt.title('Boxplot of total bill per day')
plt.show()
```

```python
# Creating Boxplot with horizontal orientation
sns.boxplot(x="total_bill", y="day", data=df, palette='rainbow', orient='h')

plt.title('Boxplot of total bill per day')
plt.show()
```

```python
# We can make it to entire dataframe too
sns.boxplot(data=df, palette='rainbow', orient='h')

plt.title('Boxplot of total bill per day')
plt.show()
```

```python
# We can use 'hue' to divide between smokers and non-smokers
sns.boxplot(x="day", y="total_bill", hue="smoker", data=df, palette='coolwarm')

plt.title('Boxplot of total bill per day')
plt.show()
```

## **Violin plot**

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
# Creating violin plot
sns.violinplot(x="day", y="total_bill", data=df, palette='rainbow')

plt.title('Violin plot of total bill per day')
plt.show()
```

```python
# Violin plot of total bill based on sex
sns.violinplot(x="day", y="total_bill", hue="sex", data=df, palette='Set1')

plt.title('Violin plot of total bill per day')
plt.show()
```

```python
# Add split = True
sns.violinplot(x="day", y="total_bill", hue="sex", data=df, palette='Set1', split=True)

plt.title('Violin plot of total bill per day')
plt.show()
```

## **Stripplot**
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.



```python
# Creating stripplot
sns.stripplot(x="day", y="total_bill", data=df, jitter=False)

plt.title('Stripplot of total bill per day')
plt.show()
```

```python
# Creating stripplot # Jitter = True
sns.stripplot(x="day", y="total_bill", data=df, palette='coolwarm', jitter=True)

plt.title('Stripplot of total bill per day')
plt.show()
```

```python
# Creating stripplot # divided by 'sex'
sns.stripplot(x="day", y="total_bill", data=df, jitter=True, hue='sex', palette='Set1')

plt.title('Stripplot of total bill per day')
plt.show()
```

```python
# Creating stripplot # dodge = True
sns.stripplot(x="day", y="total_bill", data=df,jitter=True, hue='sex',palette='Set1',  dodge=True)

plt.title('Stripplot of total bill per day (male and female)')
plt.show()
```

## **Swarmplot**

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
# Creating swarmplot
sns.swarmplot(x="day", y="total_bill", palette='Set2', data=df)

plt.title('Swarmplot of total bill per day')
plt.show()
```

```python
# Creating swarmplot split=True
sns.swarmplot(x="day", y="total_bill", data=df, hue='sex', palette='Set1')

plt.title('Swarmplot of total bill per day')
plt.tight_layout()
plt.show()
```

```python
# Creating swarmplot split=True
sns.swarmplot(x="day", y="total_bill", data=df, hue='sex', palette='Set1')

plt.title('Swarmplot of total bill per day')
# Position the legend in the upper right corner
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0) # Adjust bbox_to_anchor and loc as needed

plt.tight_layout()
plt.show()
```

## **Combined categorical plots**

Somehow, we can combine categorical plots

```python
# Combined categorical plots
sns.violinplot(x="tip", y="day", data=df,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=df,color='black',size=3)

plt.title('Combined plot of tip per day')
plt.show()
```

## **Catplot**

Catplot is the most general form of a categorical plot. It can take in a **kind** parameter to adjust the plot type:

```python
# Factor plot
sns.catplot(x='sex',y='total_bill',data=df,kind='bar', palette='rainbow')

plt.title('Catplot of average total bill by sex')
plt.show()
```
