# **Style and Color**
___

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally

**Importing seaborn and required libraries**

```python
# Importing seaborn library
import seaborn as sns

# Importing matplotlib library (supporting display)
import matplotlib.pyplot as plt
%matplotlib inline

# Customizing with 'ggplot' style
plt.style.use('ggplot')

# Importing warnings
import warnings
warnings.filterwarnings('ignore')
```

**Loading dataset**

```python
# Loading dataset
df = sns.load_dataset('tips')
```

```python
# Quick review
df.head()
```

## **Styles**

We can set particular styles:

```python
# Counting plot
sns.countplot(x='sex', data=df, palette='rainbow')

plt.show()
```

```python
# Counting plot

# Setting style using 'white'
sns.set_style('white')

# Counting plot
sns.countplot(x='sex', data=df, palette='rainbow')

plt.show()
```

```python
# Counting plot with ticks

# Adding style
sns.set_style('ticks') # Adding ticks to axes

# Counting plot
sns.countplot(x='sex', data=df, palette='rainbow')

plt.show()
```

## **Spine Removal**

```python
# Creating countplot with spine removal
sns.countplot(x='sex',data=df,palette='rainbow')
sns.despine() # spine removal

plt.show()
```

```python
# Creating countplot with both spine removal
sns.countplot(x='sex',data=df,palette='rainbow')
sns.despine(left=True) # spine removal

plt.show()
```

## **Size and Aspect**

You can use matplotlib's **plt.figure(figsize=(width,height)** to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:

```python
# Non Grid Plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=df, palette='rainbow')

plt.show()
```

```python
# Customized Grid Type Plot
sns.lmplot(x='total_bill', y='tip', height=2, aspect=4, data=df)

plt.show()
```

## **Scale and Context**

The set_context() allows you to override default parameters:

```python
# Setting context and scale
sns.set_context('poster',font_scale=0.75)
sns.countplot(x='sex',data=df, palette='coolwarm')

plt.show()
```
