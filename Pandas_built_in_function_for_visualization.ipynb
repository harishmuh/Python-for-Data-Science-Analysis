# **Pandas Built-in Data Visualization**
___


In this lecture we will learn about pandas built-in capabilities for data visualization! It's built-off of matplotlib, but it baked into pandas for easier usage!  

Let's take a look!

**Importing library**

```python
# Importing pandas
import pandas as pd

# Importing numpy
import numpy as np

# Importing matplotlib library (supporting display)
%matplotlib inline

# Importing warnings
import warnings
warnings.filterwarnings('ignore')
```

**Dataset**

We are using dummy dataset for this tutorial.

```python
# Loading dataset
url1 = 'https://raw.githubusercontent.com/harishmuh/Python-for-Data-Science-Analysis/refs/heads/main/datasets/df1_pandas_visual.csv'
url2 = 'https://raw.githubusercontent.com/harishmuh/Python-for-Data-Science-Analysis/refs/heads/main/datasets/df2_pandas_visual.csv'

df1 = pd.read_csv(url1,index_col=0)
df2 = pd.read_csv(url2)
```

**Quick review**

```python
# Quick review
display(df1.head(), df2.head())
```

```python
# Data summary
display(df1.info(), df2.info())
```

```python
# Changing data type

# Resetting the index to make the first column a regular column
df1 = df1.reset_index()

# Converting the 'index' column to datetime
df1['index'] = pd.to_datetime(df1['index'])

# Setting the 'index' column back as the index
df1 = df1.set_index('index')
```

```python
# Sanity check
print('the data type of first column of df1:', type(df1.index[0]))
```

## **Style Sheets**

Matplotlib has [style sheets](http://matplotlib.org/gallery.html#style_sheets) you can use to make your plots look a little nicer. These style sheets include plot_bmh,plot_fivethirtyeight,plot_ggplot and more. They basically create a set of style rules that your plots follow. I recommend using them, they make all your plots have the same look and feel more professional. You can even create your own if you want your company's plots to all have the same look (it is a bit tedious to create on though).

Here is how to use them.

**Before plt.style.use() your plots look like this:**

```python
df1['A'].hist()
```

Using the ggplot style

```python
# Customizing with 'ggplot' style
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Now the plots look like this

```python
df1['A'].hist()
plt.show() # Displaying without mentioning 'axes'
```

Using the 'bmh' style

```python
plt.style.use('bmh')
df1['A'].hist()
plt.show()
```

Using the 'dark background' style

```python
# Setting style: dark background
plt.style.use('dark_background')
df1['A'].hist()
plt.show()
```

Using the 'fivethirtyeight' style

```python
# Setting style:'fivethirtyeight'
plt.style.use('fivethirtyeight')
df1['A'].hist()
plt.show()
```

```python
plt.style.use('ggplot')
```

## **Plot Types**

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box', 'barh', etc..)


### **Area plot**

```python
# Area plot
df2.plot.area(alpha=0.4)
plt.show()
```

### **Barplots**

```python
df2.head()
plt.show()
```

```python
# Bar plot
df2.plot.bar()
plt.show()
```

```python
# Bar plot # Horizontal bar
df2.plot(kind='barh')
plt.show()
```

```python
# Bar plot
df2.plot(kind='box')
plt.show()
```

```python
# Stacked bar
df2.plot.bar(stacked=True)
plt.show()
```

### **Histograms**

```python
# Histograms
df1['A'].plot.hist(bins=50)
plt.show()
```

### **Line Plots**

```python
# Plotting all columns against the index
df1.plot(figsize=(12, 3), lw=1)
plt.show()
```

### **Scatter plots**

```python
# Scatter plots
df1.plot.scatter(x='A',y='B')
plt.show()
```

You can use c to color based off another column value
Use cmap to indicate colormap to use.
For all the colormaps, check out: http://matplotlib.org/users/colormaps.html

```python
# Scatter plots customized
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
plt.show()
```

Or use s to indicate size based off another column. s parameter needs to be an array, not just the name of a column:

```python
# Increasing the size of points
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
plt.show()
```

### **Boxplots**

```python
# Box plot
df2.plot.box() # Can also pass a by = argument for groupby
plt.show()
```

### **Hexagonal bin plot**

This plot can be use for Bivariate Data as alternative to scatterplot.

```python
# Hexagonal bin plot
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
plt.show()
```

### **Kernel density estimation (KDE) plot**

```python
# KDE
df2['a'].plot.kde()
plt.show()
```

```python
df2.plot.density()
plt.show()
```

That's it! Hopefully you can see why this method of plotting will be a lot easier to use than full-on matplotlib, it balances ease of use with control over the figure. A lot of the plot calls also accept additional arguments of their parent matplotlib plt. call.
