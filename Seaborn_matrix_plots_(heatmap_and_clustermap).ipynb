# **Matrix Plots**
___

Matrix plots allow us to plot data as color-encoded matrices and can also be used to indicate clusters within the data (We will use that for clustering data in machine learning section).

Let's begin by exploring seaborn's heatmap and clustermap:

**Importing seaborn**

```python
# Importing seaborn library
import seaborn as sns

# Importing matplotlib library (supporting display)
import matplotlib.pyplot as plt
%matplotlib inline

# Importing warnings (remove warning)
import warnings
warnings.filterwarnings('ignore')
```

**Loading dataset**

We will use two datasets 'tips' and 'flights', both are available in seaborn library

```python
# Loading dataset
df_flights = sns.load_dataset('flights') # Available in seaborn
df_tips = sns.load_dataset('tips') # Available in seaborn
```

```python
# Displaying data
print('Quick review of df_flight')
display(df_flights.head())
print('')
print('Quick review of df_tips')
display(df_tips.head())
```

## **Heatmap**

In order for a heatmap to work properly, your data should already be in a matrix form, the sns.heatmap function basically just colors it in for you. For example:

**Tips data**

We will make correlation matrix and heatmap based on tips data

```python
# Matrix form for correlation data of df_tips
df_tips.corr(numeric_only=True)
```

```python
# Creating Heatmap correlation matrix
sns.heatmap(df_tips.corr(numeric_only=True))

# Display
plt.title('Correlation heatmap of tips dataframe')
plt.show()
```

```python
# Correlation matrix with annotation
sns.heatmap(df_tips.corr(numeric_only=True), annot=True)

# Display
plt.title('Correlation heatmap of tips dataframe')
plt.show()
```

**Flights data**

We will make pivot table and heatmap based on flights data

```python
# Creating pivot table
df_flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
# Creating heatmap using pivot table
pivot_table = df_flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(pivot_table)

# Display
plt.title('Heatmap of flights pivot table')
plt.show()
```

On the heatmap above, we can see the higher number of the passenggers on the lighter the colour. The highest number of customer was on July, 1960.

```python
# Customizing pivot table # Line color white
sns.heatmap(data = pivot_table,cmap='magma',linecolor='white',linewidths=1)

# Display
plt.title('Heatmap of flights pivot table')
plt.show()
```

## **Cluster map**

The clustermap uses hierarchal clustering to produce a clustered version of the heatmap. For example:

```python
# Creating cluster map
sns.clustermap(pivot_table)

# Display
plt.show()
```

```python
# Creating the clustermap
cluster = sns.clustermap(pivot_table)

# Add centered title using suptitle
cluster.fig.suptitle('Cluster Map of Flights DataFrame', fontsize=14, y=1.05)
# y=1.05  moves the title upward above the heatmap to avoid overlapping.

plt.show()
```

Can you notice now how the years and months are no longer in order? instead they are grouped by similarity in value (passenger count). That means we can begin to infer things from this plot, such as August and July being similar (makes sense, since they are both summer travel months)

```python
# More options to get the information a little clearer like normalization
sns.clustermap(data=pivot_table,cmap='coolwarm',standard_scale=1)

# Add centered title using suptitle
cluster.fig.suptitle('Cluster Map of Flights DataFrame', fontsize=14, y=1.05)

# Display
plt.show()
```
