# **Error Handling**
---

## **Try-Except**

```python
# Normal without error
a = 10
b = 5
# try: Trying codeline
try:
  result = a/b
  print(result)
# Except: will be executed if facing an error.
except ZeroDivisionError:
  print('Error: Division by zero')
```

```python
# Zero division error
a = 10
b = 0
try:
  result = a/b
  print(result)
except ZeroDivisionError:
  print('Error: Division by zero')
```

## **Multiple Except**

```python
# Handling multiple type of errors
a = 0
b = 'a'
try:
  result = a/b
  print(result)
except ZeroDivisionError:
  print('Error: Division by zero')
except TypeError:
  print('Error: Type error')
```

## **Try - Except - Else**

```python
# Handling multiple type of errors (with else)
a = 10
b = 5
try:
  result = a/b
  print(result)
except TypeError:
  print('Error: Type error')
except ZeroDivisionError:
  print('Error: Division by zero')
except:
  print('Error: Unknown error')
else:
  print('Operation is succeed')
```

```python
# Handling multiple type of errors (with else)
# simulation
a = 10
b = 'c'
try:
  result = a/b
  print(result)
except TypeError:
  print('Error: Type error')
except ZeroDivisionError:
  print('Error: Division by zero')
except:
  print('Error: Unknown error')
else:
  print('Operation is succeed')
```
