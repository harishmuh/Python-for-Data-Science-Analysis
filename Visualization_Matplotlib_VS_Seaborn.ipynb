# **Visualization**

This tutorial describes how to visualize graphs from iris plant dataset using  Matplotlib and Seaborn  

```python
# Importing packages and libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Custom display
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading files

uci_dataset = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

df = pd.read_csv(uci_dataset)
```

```python
# Adding column name

df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
# Quick review

display(df.head(), df.tail())
```

## Matplotlib

```python
# Scatter plot # Matplotlib
df.plot.scatter(x='sepal_length', y='sepal_width')
plt.xlabel('sepal_length (cm)')
plt.ylabel('sepal_width (cm)')
plt.title('Scatter plot of sepal length and sepal width of Iris plant species')
plt.show()
```

```python
# Box plot
# Boxplot of comparison between parameters
plt.boxplot(df.drop('species', axis=1))
plt.xticks(range(1,5), df.drop('species', axis=1).columns) # Changed to use columns for labels
plt.show()
```

```python
# Violin plot
# Violin plot of comparison between parameters
plt.violinplot(df.drop('species', axis=1))
plt.xticks(range(1,5), df.drop('species', axis=1).columns)
plt.show()
```

```python
# boxplot of Iris plants parameters
plt.figure(figsize = (10,10))
df.boxplot(column=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], by='species')
plt.tight_layout()
plt.show()
```

```python
# Scatter matrix
pd.plotting.scatter_matrix(df.drop('species', axis=1), alpha=0.2, figsize=(10,10))
plt.show()
```

## Seaborn Library

```python
# Scatterplot # Seaborn
sns.scatterplot(data=df, x='sepal_length', y='sepal_width')
plt.title('Scatter plot of sepal length and sepal width of Iris plants')
plt.xlabel('sepal_length (cm)')
plt.ylabel('sepal_width (cm)')
plt.tight_layout()
plt.show()
```

```python
# Scatterplot # Seaborn
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue= 'species')
plt.title('Scatter plot of sepal length and sepal width of Iris plant species')
plt.xlabel('sepal_length (cm)')
plt.ylabel('sepal_width (cm)')
plt.tight_layout()
plt.show()
```

```python
# Joint plot
sns.jointplot(data=df, x='sepal_length', y='sepal_width')
plt.title('Joint plot of sepal length and sepal width of Iris plants')
plt.xlabel('sepal_length (cm)')
plt.ylabel('sepal_width (cm)')
plt.tight_layout()
plt.show()
```

```python
# Joint plot
sns.jointplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Joint plot of sepal length and sepal width of Iris plant species')
plt.xlabel('sepal_length (cm)')
plt.ylabel('sepal_width (cm)')
plt.tight_layout()
plt.show()
```

```python
# Facetgrid
sns.FacetGrid(df, hue='species').map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()
plt.title('Facetgrid of sepal length and sepal width of Iris plant species')
plt.xlabel('sepal_length (cm)')
plt.ylabel('sepal_width (cm)')
plt.tight_layout()
plt.show()
```

```python
# Facetgrid
g = sns.FacetGrid(df, hue='species', height=5, aspect=1.2)  # Set height and aspect ratio
g.map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()
plt.title('Facetgrid of sepal length and sepal width of Iris plant species')
plt.xlabel('sepal_length (cm)')
plt.ylabel('sepal_width (cm)')
plt.tight_layout()
plt.show()
```

```python
# Boxplot # Seaborn
sns.boxplot(data=df, x='species', y='sepal_length', palette='husl')
plt.title('Boxplot of sepal length of Iris plant species')
plt.xlabel('species')
plt.ylabel('sepal_length (cm)')
plt.tight_layout()
plt.show()
```

```python
# Violin plot
sns.violinplot(data=df, x='species', y='sepal_length', palette='husl')
plt.title('Violin plot of sepal length of Iris plant species')
plt.xlabel('species')
plt.ylabel('sepal_length (cm)')
plt.show()
```

```python
# Swarm plot
sns.swarmplot(data=df, x='species', y='sepal_length', palette='husl')
plt.title('Swarm plot of sepal length of Iris plant species')
plt.xlabel('species')
plt.ylabel('sepal_length (cm)')
plt.tight_layout()
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue='species', palette='husl')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to make space for the legend
plt.show()
```
