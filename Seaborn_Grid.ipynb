# Grids
___

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

**Imporing seaborn and required libraries**

```python
# Importing seaborn library
import seaborn as sns

# Importing matplotlib library (supporting display)
import matplotlib.pyplot as plt
%matplotlib inline

# Customizing with 'ggplot' style
plt.style.use('ggplot')

# Importing warnings (remove warning)
import warnings
warnings.filterwarnings('ignore')
```

**Loading dataset**

```python
# Loading dataset
df = sns.load_dataset('iris') # Available in seaborn
```

```python
# Quick review
df.head()
```

## **PairGrid**

Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
# Grid only
sns.PairGrid(df)

# Display
plt.show()
```

```python
# Mapping scatter plot to the grid
g = sns.PairGrid(df)
g.map(plt.scatter)

# Display
plt.show()
```

We can map variety of plots to the grid

```python
# Mapping plot to upper, diagonal, and lower grid
g = sns.PairGrid(df)
g.map_diag(plt.hist) # Mapping histogram plot to the diagonal grid
g.map_upper(plt.scatter) # Mapping scatter plot to the upper grid
g.map_lower(sns.kdeplot) # Mapping kde plot to the lower grid

# Display
plt.show()
```

## **Pairplot**

pairplot is a simpler version of PairGrid (you'll use quite often)

```python
# Creating pairplot
sns.pairplot(df)

# Display
plt.show()
```

```python
# More customized pairplot
sns.pairplot(df, hue='species', palette='rainbow')

# Display
plt.show()
```

## **Facet Grid**

FacetGrid is the general way to create grids of plots based off of a feature. We will use 'tips' dataset.

```python
# Loading dataset
df_tips = sns.load_dataset('tips')
```

```python
# Quick review
df_tips.head()
```

```python
# Only the grid
g = sns.FacetGrid(df_tips, col='time', row='smoker')

# Display
plt.show()
```

```python
# Creating histograms on the FacetGrid
g = sns.FacetGrid(df_tips, col='time', row='smoker')
g.map(plt.hist, 'total_bill')

# Display
plt.show()
```

```python
# Creating scatter plots on the FacetGrid
g = sns.FacetGrid(df_tips, col='time', hue='sex', row='smoker')
g = g.map(plt.scatter, 'total_bill', 'tip').add_legend()

# Display
plt.show()
```

## **JointGrid**

JointGrid is the general version for jointplot() type grids, for a quick example:



```python
# JointGrid only
g = sns.JointGrid(x='total_bill', y='tip', data=df_tips)

# Display
plt.show()
```

```python
# Mapping regression plot to the JointGrid
g = sns.JointGrid(x='total_bill', y='tip', data=df_tips)
g = g.plot(sns.regplot, sns.distplot)

# Display
plt.show()
```
